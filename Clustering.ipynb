**Clustering**

Q1.What is unsupervised learning in the context of machine learning?

-  Unsupervised learning is a type of machine learning where algorithms analyze and find hidden patterns, structures, or relationships in unlabeled data, without any pre-existing categories or human guidance, essentially teaching themselves to group or understand data based on inherent similarities.

Q2.How does K-Means clustering algorithm work?

-  K-means clustering works by iteratively grouping similar data points into a predefined number (K) of clusters, finding centers (centroids) for each, and reassigning points to the nearest centroid until the cluster assignments stabilize.

Q3.Explain the concept of a dendrogram in hierarchical clustering.

-  A dendrogram in hierarchical clustering is a tree-like diagram that visually represents the nested grouping of data points, showing how individual items merge into clusters and how those clusters combine, with the height of branches indicating the distance or dissimilarity between them.

Q4.What is the main difference between K-Means and Hierarchical Clustering?

-  The main difference is that K-Means requires specifying the number of clusters (\(K\)) in advance and efficiently partitions data into non-overlapping groups, whereas hierarchical clustering does not require a predefined \(K\), building a nested tree structure (dendrogram) instead.

Q5.What are the advantages of DBSCAN over K-Means?

-  DBSCAN (Density-Based Spatial Clustering of Applications with Noise) offers significant advantages over K-Means, primarily by identifying arbitrary, non-spherical cluster shapes and handling noise/outliers without requiring the pre-specification of cluster numbers (\(k\)).

Q6.When would you use Silhouette Score in clustering?

-  The Silhouette Score is used in unsupervised learning to evaluate the quality of clustering, specifically to determine the optimal number of clusters (\(k\)) and to validate how well data points are separated and grouped.

Q7.What are the limitations of Hierarchical Clustering?

-   Key limitations include sensitivity to noise and outliers, irreversibility of merges (once merged, they cannot be undone), and difficulty in interpreting the resulting dendrograms.

Q8.Why is feature scaling important in clustering algorithms like K-Means?

-  Feature scaling is crucial in K-Means because the algorithm relies on Euclidean distance to determine similarity between data points.

Q9.How does DBSCAN identify noise points?

-  DBSCAN identifies noise points by classifying data points with low-density neighborhoods as outliers that do not belong to any cluster.

Q10.Define inertia in the context of K-Means.

-  In the context of K-Means clustering, inertia (often called the within-cluster sum-of-squares or WCSS) is a metric that measures how internally coherent or compact the clusters are.

Q11.What is the elbow method in K-Means clustering?

-  The elbow method is a heuristic used to determine the optimal number of clusters (\(k\)) in K-means clustering by plotting the within-cluster sum of squares (WCSS) or inertia against the number of clusters.

Q12.Describe the concept of "density" in DBSCAN.

-  In DBSCAN, "density" defines clusters as continuous, high-concentration regions of data points separated by sparse (low-density) areas.

Q13.Can hierarchical clustering be used on categorical data?

-  Yes, hierarchical clustering can be used on categorical data.

Q14.What does a negative Silhouette Score indicate?

-  A negative silhouette score (ranging from -1 to 0) indicates that data points are likely assigned to the wrong cluster.

Q15.Explain the term "linkage criteria" in hierarchical clustering.

-  Linkage criteria in hierarchical clustering define the distance metric used to calculate the separation between two clusters, determining which pairs are merged at each step.

Q16.Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

-  K-Means clustering often performs poorly on data with varying cluster sizes, densities, or non-spherical shapes because its underlying optimization objective assumes that all clusters are spherical, roughly equal in size, and similar in density.

Q17.What are the core parameters in DBSCAN, and how do they influence clustering?

-  DBSCAN (Density-Based Spatial Clustering of Applications with Noise) relies on two core parameters, Epsilon and MinPts.

Q18.How does K-Means++ improve upon standard K-Means initialization?

-  K-Means++ improves upon standard K-Means by using a smart, probabilistic initialization that spreads out initial centroids, significantly reducing the likelihood of poor clustering and accelerating convergence.

Q19.What is agglomerative clustering?

-  Agglomerative clustering is a bottom-up hierarchical machine learning algorithm that groups similar data points into clusters by initially treating each point as a single cluster and iteratively merging the closest pairs.

Q20.What makes Silhouette Score a better metric than just inertia for model evaluation?

-  The Silhouette Score is superior to inertia because it provides a holistic measure of clustering quality, combining both cohesion (how close points are within a cluster) and separation (how distinct clusters are from each other).

```python
#Q25.Use make_circles to generate synthetic data and cluster it using DBSCAN. Plot the result.

import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import pandas as pd
import numpy as np

# Generate synthetic data: circles
X, y = make_circles(n_samples=500, factor=0.5, noise=0.05, random_state=42)

# Standardize the features
X_scaled = StandardScaler().fit_transform(X)

# Apply DBSCAN
# eps: The maximum distance between two samples for one to be considered as in the neighborhood of the other.
# min_samples: The number of samples in a neighborhood for a point to be considered as a core point.
dbscan = DBSCAN(eps=0.3, min_samples=5)
clusters = dbscan.fit_predict(X_scaled)

# Plotting the result
plt.figure(figsize=(8, 6))
unique_labels = set(clusters)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]

for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = [0, 0, 0, 1]

    class_member_mask = (clusters == k)
    xy = X_scaled[class_member_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=6, label=f'Cluster {k}' if k != -1 else 'Noise')

plt.title('DBSCAN Clustering of make_circles Data')
plt.xlabel('Feature 1 (scaled)')
plt.ylabel('Feature 2 (scaled)')
plt.legend(loc='upper right')
plt.grid(True, linestyle='--', alpha=0.6)
plt.savefig('dbscan_circles.png')

# Print cluster counts
cluster_counts = pd.Series(clusters).value_counts().sort_index()
print("Cluster Distribution:")
print(cluster_counts)
```

```python
#Q27.Generate synthetic data using make_blobs with varying cluster standard deviations and cluster with DBSCAN.

import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import numpy as np
import pandas as pd

# 1. Generate synthetic data with varying cluster standard deviations
centers = [[-5, -5], [0, 0], [5, 5]]
cluster_std = [0.5, 2.5, 1.0]  # Varying standard deviations
X, y = make_blobs(n_samples=1000, centers=centers, cluster_std=cluster_std, random_state=42)

# 2. Standardize the features
X_scaled = StandardScaler().fit_transform(X)

# 3. Apply DBSCAN
# We choose parameters that aim to capture the different densities,
# though DBSCAN often struggles with varying densities.
dbscan = DBSCAN(eps=0.3, min_samples=10)
clusters = dbscan.fit_predict(X_scaled)

# 4. Plot the results
plt.figure(figsize=(10, 7))

unique_labels = set(clusters)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]

for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = [0, 0, 0, 1]

    class_member_mask = (clusters == k)
    xy = X_scaled[class_member_mask]
    plt.scatter(xy[:, 0], xy[:, 1], s=30, color=tuple(col), edgecolors='k',
                label=f'Cluster {k}' if k != -1 else 'Noise', alpha=0.7)

plt.title('DBSCAN Clustering on Blobs with Varying Standard Deviations')
plt.xlabel('Feature 1 (scaled)')
plt.ylabel('Feature 2 (scaled)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.savefig('dbscan_varying_blobs.png')

# Print cluster counts
cluster_counts = pd.Series(clusters).value_counts().sort_index()
print("Cluster Distribution (including -1 for Noise):")
print(cluster_counts)
```

```python
#Q28.Load the Digits dataset, reduce it to 2D using PCA, and visualize clusters from K-Means.

import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load the Digits dataset
digits = load_digits()
X = digits.data
y = digits.target

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Reduce dimensions to 2D using PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Apply K-Means clustering
# Since we know there are 10 digits (0-9), we'll use n_clusters=10
kmeans = KMeans(n_clusters=10, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_pca)

# Plotting
plt.figure(figsize=(10, 8))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='tab10', alpha=0.6, edgecolors='k')
plt.colorbar(scatter, label='Cluster ID')

# Plot the centroids
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=200, linewidths=3, color='black', zorder=10, label='Centroids')

plt.title('K-Means Clustering on Digits Dataset (PCA-reduced 2D Projection)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.savefig('digits_pca_kmeans.png')

print("PCA explained variance ratio:", pca.explained_variance_ratio_)
```

```python
#Q29.Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.

import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import pandas as pd

# 1. Generate synthetic data
X, y = make_blobs(n_samples=500, centers=4, cluster_std=1.0, random_state=42)

# 2. Evaluate silhouette scores for k = 2 to 5
k_values = [2, 3, 4, 5]
silhouette_scores = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(X)
    score = silhouette_score(X, cluster_labels)
    silhouette_scores.append(score)

# 3. Create a DataFrame for easy plotting
results_df = pd.DataFrame({
    'k': k_values,
    'Silhouette Score': silhouette_scores
})

# 4. Display as a bar chart
plt.figure(figsize=(8, 6))
bars = plt.bar(results_df['k'].astype(str), results_df['Silhouette Score'], color='skyblue', edgecolor='navy')

# Add labels and title
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for different k values')
plt.ylim(0, 1.0) # Silhouette score ranges from -1 to 1, but usually positive for blobs
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add value labels on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.02, round(yval, 3), ha='center', va='bottom')

plt.savefig('silhouette_scores_bar.png')

print("Silhouette Scores:")
print(results_df)
```

```python
#Q30.Load the Iris dataset and use hierarchical clustering to group data. Plot a dendrogram with average linkage.

import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import dendrogram, linkage
import pandas as pd

# Load the Iris dataset
iris = load_iris()
X = iris.data
feature_names = iris.feature_names

# Perform hierarchical clustering using 'average' linkage
# Linkage methods: 'single', 'complete', 'average', 'ward'
Z = linkage(X, method='average')

# Plot the dendrogram
plt.figure(figsize=(12, 7))
dendrogram(
    Z,
    labels=iris.target, # Labeling with actual species to see grouping quality
    leaf_rotation=90.,
    leaf_font_size=8.,
)

plt.title('Hierarchical Clustering Dendrogram (Iris Dataset - Average Linkage)')
plt.xlabel('Sample Index (Color-coded by Species)')
plt.ylabel('Distance (Average Linkage)')
plt.axhline(y=1.8, color='r', linestyle='--') # Example cut-off line
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.savefig('iris_dendrogram.png')

print("Linkage matrix shape:", Z.shape)
```

```python
#Q31.Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# 1. Generate synthetic data with overlapping clusters
# Using a higher cluster_std to ensure they overlap
X, y = make_blobs(n_samples=500, centers=3, cluster_std=2.5, random_state=42)

# 2. Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans.fit(X)

# 3. Create a meshgrid for decision boundaries
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))

# Predict across the meshgrid
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# 4. Plotting
plt.figure(figsize=(10, 8))

# Plot decision boundaries
plt.contourf(xx, yy, Z, alpha=0.3, cmap='viridis')

# Plot the data points
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, s=30, edgecolor='k', cmap='viridis', alpha=0.7)

# Plot the centroids
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=200,
            linewidths=3, color='red', zorder=10, label='Centroids')

plt.title('K-Means Decision Boundaries on Overlapping Clusters')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.savefig('kmeans_decision_boundaries.png')

print("Cluster centroids:")
print(centroids)
```

```python
#Q32.Load the Digits dataset and apply DBSCAN after reducing dimensions with t-SNE. Visualize the results.

import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np

# 1. Load the Digits dataset
digits = load_digits()
X = digits.data
y = digits.target

# 2. Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Reduce dimensions with t-SNE
# t-SNE is generally better at preserving local structures for visualization than PCA.
tsne = TSNE(n_components=2, perplexity=30, random_state=42, init='pca', learning_rate='auto')
X_tsne = tsne.fit_transform(X_scaled)

# 4. Apply DBSCAN on the t-SNE output
# t-SNE creates dense "islands", so we tune DBSCAN parameters for this 2D space
dbscan = DBSCAN(eps=2.5, min_samples=5)
clusters = dbscan.fit_predict(X_tsne)

# 5. Visualize the results
plt.figure(figsize=(10, 8))

unique_labels = set(clusters)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]

for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = [0, 0, 0, 1]

    class_member_mask = (clusters == k)
    xy = X_tsne[class_member_mask]
    plt.scatter(xy[:, 0], xy[:, 1], s=30, color=tuple(col),
                label=f'Cluster {k}' if k != -1 else 'Noise',
                alpha=0.6, edgecolors='k')

plt.title('DBSCAN Clustering on Digits (t-SNE Reduced 2D Space)')
plt.xlabel('t-SNE feature 1')
plt.ylabel('t-SNE feature 2')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('digits_tsne_dbscan.png')

# Print cluster distribution
import pandas as pd
print("Cluster Distribution:")
print(pd.Series(clusters).value_counts().sort_index())
```

```python
#Q34.Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a line plot.

import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Load the Breast Cancer dataset
data = load_breast_cancer()
X = data.data

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Calculate inertia for K = 2 to 6
k_values = range(2, 7)
inertia_values = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia_values.append(kmeans.inertia_)

# Plot the results
plt.figure(figsize=(8, 6))
plt.plot(k_values, inertia_values, marker='o', linestyle='-', color='b')
plt.title('Elbow Method: Inertia vs Number of Clusters (Breast Cancer Dataset)')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.xticks(k_values)
plt.grid(True, linestyle='--', alpha=0.7)
plt.savefig('breast_cancer_inertia.png')

# Print inertia values
results = pd.DataFrame({'K': k_values, 'Inertia': inertia_values})
print(results)
```

```python
#Q40.Load the Digits dataset, apply PCA to reduce to 3 components, then use KMeans and visualize with a 3D scatter plot.

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load the Digits dataset
digits = load_digits()
X = digits.data
y = digits.target

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Reduce dimensions to 3 using PCA
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_scaled)

# Apply KMeans clustering (10 clusters for 10 digits)
kmeans = KMeans(n_clusters=10, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_pca)

# Visualize with a 3D scatter plot
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot using the cluster labels for color
scatter = ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2],
                    c=clusters, cmap='tab10', alpha=0.6, edgecolors='w', s=40)

# Plot the centroids
centroids = kmeans.cluster_centers_
ax.scatter(centroids[:, 0], centroids[:, 1], centroids[:, 2],
           marker='x', s=200, color='black', linewidths=3, label='Centroids')

ax.set_title('3D PCA Projection of Digits Dataset with KMeans Clustering')
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')

# Add colorbar
fig.colorbar(scatter, ax=ax, label='Cluster ID', shrink=0.5, aspect=5)

plt.legend()
plt.tight_layout()
plt.savefig('digits_pca_3d.png')

print("PCA explained variance ratio for 3 components:", pca.explained_variance_ratio_)
print("Total variance explained:", sum(pca.explained_variance_ratio_))
```

```python
#Q41.Generate synthetic blobs with 5 centers and apply KMeans. Then use silhouette_score to evaluate the clustering.

import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# 1. Generate synthetic blobs with 5 centers
X, y = make_blobs(n_samples=600, centers=5, cluster_std=0.8, random_state=42)

# 2. Apply K-Means with k=5
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(X)

# 3. Calculate the Silhouette Score
score = silhouette_score(X, cluster_labels)

# 4. Visualization
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=cluster_labels, s=30, cmap='viridis', alpha=0.7, edgecolors='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            marker='X', s=200, color='red', label='Centroids')

plt.title(f'K-Means Clustering (k=5)\nSilhouette Score: {score:.4f}')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.savefig('kmeans_5_blobs.png')

print(f"Silhouette Score for k=5: {score:.4f}")
```

```python
#Q42.Load the Breast Cancer dataset, reduce dimensionality using PCA, and apply Agglomerative Clustering. Visualize in 2D.

import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler

# 1. Load the Breast Cancer dataset
data = load_breast_cancer()
X = data.data
y = data.target

# 2. Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Reduce dimensionality to 2D using PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# 4. Apply Agglomerative Clustering
# We'll use 2 clusters to see how it aligns with the Benign/Malignant labels
agg_clustering = AgglomerativeClustering(n_clusters=2)
clusters = agg_clustering.fit_predict(X_pca)

# 5. Visualize the results
plt.figure(figsize=(10, 7))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='coolwarm', alpha=0.7, edgecolors='k')

# Create a legend for the clusters
plt.title('Agglomerative Clustering on Breast Cancer Dataset (PCA-reduced)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True, linestyle='--', alpha=0.5)

# Add a color bar
cb = plt.colorbar(scatter)
cb.set_label('Cluster ID')

plt.savefig('breast_cancer_agg_pca.png')

print("PCA Variance Ratios:", pca.explained_variance_ratio_)
print("Total Variance Explained:", sum(pca.explained_variance_ratio_))
```

```python
#Q46.Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4 features).

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.cluster import KMeans

# Load Wine dataset
wine = load_wine()
X = wine.data
feature_names = wine.feature_names

# Create DataFrame with first 4 features
df = pd.DataFrame(X[:, :4], columns=feature_names[:4])

# Apply KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
df["Cluster"] = kmeans.fit_predict(X)

# Plot seaborn pairplot
sns.pairplot(df, hue="Cluster", diag_kind="kde", palette="Set1")
plt.suptitle("KMeans Clusters on Wine Dataset (First 4 Features)", y=1.02)
plt.show()
```

```python
#Q48.Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the clusters.

import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler

# 1. Load the Digits dataset
digits = load_digits()
X = digits.data
y = digits.target

# 2. Standardize the data
# t-SNE is sensitive to scale
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Apply t-SNE to reduce to 2 components
tsne = TSNE(n_components=2, perplexity=30, random_state=42, init='pca', learning_rate='auto')
X_tsne = tsne.fit_transform(X_scaled)

# 4. Apply Agglomerative Clustering
# We'll use 10 clusters to represent the 10 digits (0-9)
agg = AgglomerativeClustering(n_clusters=10, linkage='ward')
labels = agg.fit_predict(X_tsne)

# 5. Visualize the clusters
plt.figure(figsize=(10, 8))
scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels, cmap='tab10', s=15, alpha=0.8)

plt.colorbar(scatter, label='Cluster ID')
plt.title('Agglomerative Clustering on Digits (t-SNE Reduced Space)')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.grid(True, linestyle='--', alpha=0.5)

plt.savefig('digits_tsne_agg.png')

print("Clustering distribution:")
import pandas as pd
print(pd.Series(labels).value_counts().sort_index())
```
